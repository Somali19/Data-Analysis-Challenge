Correlation Matrix with Heatmap

```
# Heatmap for visualizing correlation analysis
cor = df.corr()
plt.figure(figsize=(18,12))
sns.heatmap(cor,annot=True,cmap='coolwarm')  
plt.show()
```

```
cor[abs(cor['price']) > 0.1]['price']
```

price  - normalized-losses , price - height , price - stroke and price - compression-ratio have weak positive linear relationship.

price - bore have moderate positive linear relationship.

price - wheel-base , price - length , price - width , price - curb-weight , price - engine-size  ,  price - horsepower have strong relationship
